### Análise de Dados - Tecnologia Embarcada (.shp)

```python
import pandas as pd
import shapefile
from shapely.geometry import LineString, Point, MultiLineString, Polygon
import matplotlib.pyplot as plt
from descartes.patch import PolygonPatch
%matplotlib inline
```

```python
json_hv = pd.read_json("data\sample_data_MyJohnDeere_Harvest\doc\Merriweather Farms-JT-01-Soybeans-Deere-Metadata.json", orient = 'index')
json_hv
```

```python
json_hv[0]['DataAttributes']
```

```python
farm = json_hv[0]['FarmName']
field_name = json_hv[0]['FieldName']
crop_season = json_hv[0]['CropSeason']
crop_token = json_hv[0]['CropToken']
```

```python
#Analisando Dados de Havister
sf = shapefile.Reader("data\sample_data_MyJohnDeere_Harvest\doc\Merriweather Farms-JT-01-Soybeans.shp")

shapes = sf.shapes()
head = [x[0] for x in sf.fields][1:]  #cabeçalhos
records = [y[:] for y in sf.records()]  #atributos
lat = [s.points[0][0] for s in sf.shapes()]  #Latitude
long = [s.points[0][1] for s in sf.shapes()] #Longtude

sf_df = pd.DataFrame(data = records, columns= head)

sf_df['Lat'] = lat
sf_df['Long'] = long
```

```python
sf_df['DISTANCE'] = sf_df['DISTANCE']*0.3048 #m
sf_df['SWATHWIDTH'] = sf_df['SWATHWIDTH']*0.3048 #m
sf_df['VRYIELDVOL'] = sf_df['VRYIELDVOL']*0.03524/0.404686 #m³/ha
sf_df['WetMass'] = sf_df['WetMass']*0.453592/60//0.404686 #sac/ha
sf_df['Size'] = sf_df['DISTANCE']*sf_df['SWATHWIDTH']/10000 #ha
sf_df['Elevation'] = sf_df['Elevation']*0.3048 #m
```

```python
sf_df.head()
```

```python
fig = plt.figure(figsize = (18,12))
title = "Poligono do Talhão - "+farm+" - "+field_name
ax = fig.add_subplot()
ax.set_title(title)
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')

px = sf_df['Lat']
py = sf_df['Long']
ax.scatter(px, py, color = 'black', marker = '.', s = 0.075)
```

```python
sf_df.describe()
```

```python
fig = plt.figure(figsize = (18,12))
ax = fig.subplots()
ax.hist(sf_df['WetMass'], 100)
plt.show()
```
